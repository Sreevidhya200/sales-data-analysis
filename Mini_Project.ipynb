📘 Project Summary / Introduction

This project focuses on analyzing a retail sales dataset using Python data analysis libraries such as Pandas, NumPy, and Matplotlib. The main objective of this project is to understand sales performance, profit distribution, and business trends using data analysis techniques.

In this project, Pandas is used as the main tool for data manipulation and analysis. It helps in reading the dataset, cleaning the data, performing grouping operations, filtering records, and generating statistical insights. NumPy is used for numerical operations and calculations, while Matplotlib is used for data visualization to present the analysis results in graphical form.

The dataset contains information about orders, products, categories, sales, profits, regions, and countries. By analyzing this dataset, we can identify important business insights such as which category generates the highest sales, which sub-category provides the most profit, and which regions contribute the most to company revenue.

This project demonstrates how data analysis techniques can help businesses make better decisions by understanding patterns and trends in the data.

```python
# Q1: Import required libraries for data analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_csv("/content/drive/MyDrive/SuperStoreOrders - SuperStoreOrders.csv")


```

```python
#columns shows all column names
print("Columns in dataset:")
print(df.columns)
```

```python
# Q4: Basic statistics of numerical columns
# describe() gives summary statistics like mean, min, max, std
print(df.describe())
```

```python
from collections import defaultdict
# Q6: Create a new column using NumPy
# Creating Profit Ratio column
# profit ratio = profit / sales

# Convert 'sales' column to numeric, coercing errors to NaN


df["profit_ratio"] = np.where(
    df["sales"] != 0, # Check for non-zero sales to avoid division by zero
    df["profit"] / df["sales"],
    0
)

print(df[["sales","profit","profit_ratio"]].head())
```

```python
# Q5: Check missing values in dataset
# isnull() checks missing values
# sum() counts them

print("Missing values in dataset:")
print(df.isnull().sum())

```

```python
# Q7: Find total sales of the company
total_sales = df["sales"].sum()

print("Total Sales:", total_sales)
```

```python
# Q8: Find total profit
total_profit = df["profit"].sum()

print("Total Profit:", total_profit)
```

```python
# Q9: Find average sales
avg_sales = df["sales"].mean()

print("Average Sales:", avg_sales)
```

```python
# Q10: Which category has highest sales?
category_sales = df.groupby("category")["sales"].sum()

print("Sales by Category:")
print(category_sales)
```

```python
# Q11: Which sub-category gives highest profit?
sub_profit = df.groupby("sub_category")["profit"].sum()
# sort values descending
sub_profit = sub_profit.sort_values(ascending=False)

print("Profit by Sub Category:")
print(sub_profit)
```

```python
# Q12: Top 10 products based on sales
top_products = df.sort_values(by="sales", ascending=False)

print("Top 10 Products by Sales:")
print(top_products[["product_name","sales"]].head(10))

```

```python
# Q13: Sales by Region
region_sales = df.groupby("region")["sales"].sum()

print("Sales by Region:")
print(region_sales)
```

```python
# Q14: Number of orders per category
order_count = df.groupby("category")["order_id"].count()

print("Number of Orders by Category:")
print(order_count)
```

```python
# Q15: Which country has highest sales?
country_sales = df.groupby("country")["sales"].sum()

country_sales = country_sales.sort_values(ascending=False)

print("Top Countries by Sales:")
print(country_sales.head(10))

```

```python
# Q16: Filter data (High Profit Orders)
# selecting rows where profit > 500

high_profit = df[df["profit"] > 500]

print("High Profit Orders:")
print(high_profit.head())
```

```python
# Q17: Find average profit by category
avg_profit_category = df.groupby("category")["profit"].mean()

print("Average Profit by Category:")
print(avg_profit_category)
```

```python
# Q18: Visualization – Sales by Category
category_sales.plot(kind="bar")

plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")

plt.show()
```

```python
# Q19: Visualization – Profit by Category
profit_category = df.groupby("category")["profit"].sum()

profit_category.plot(kind="bar")

plt.title("Profit by Category")
plt.xlabel("Category")
plt.ylabel("Total Profit")

plt.show()
```

```python
# Q20: Visualization – Top 10 Sub Categories by Profit
top_sub = df.groupby("sub_category")["profit"].sum()

top_sub = top_sub.sort_values(ascending=False).head(10)

top_sub.plot(kind="bar")

plt.title("Top 10 Sub Categories by Profit")
plt.xlabel("Sub Category")
plt.ylabel("Profit")

plt.show()
```

```python
# INSIGHTS FROM THE DATASET
# Q21: Which category has the highest sales?
# --------------------------------------------------------------

category_sales = df.groupby("category")["sales"].sum()

highest_sales_category = category_sales.idxmax()   # find category with highest sales

print("Category with Highest Sales:", highest_sales_category)


# --------------------------------------------------------------
# Q22: Which category gives the highest profit?
# --------------------------------------------------------------

category_profit = df.groupby("category")["profit"].sum()

highest_profit_category = category_profit.idxmax()

print("Category with Highest Profit:", highest_profit_category)


# --------------------------------------------------------------
# Q23: Which region generates the highest sales?
# --------------------------------------------------------------

region_sales = df.groupby("region")["sales"].sum()

top_region = region_sales.idxmax()

print("Region with Highest Sales:", top_region)


# --------------------------------------------------------------
# Q24: Which sub-category gives the highest profit?
# --------------------------------------------------------------

sub_profit = df.groupby("sub_category")["profit"].sum()

top_subcategory = sub_profit.idxmax()

print("Sub Category with Highest Profit:", top_subcategory)


# --------------------------------------------------------------
# Q25: Find the most sold product
# --------------------------------------------------------------

top_product = df.groupby("product_name")["sales"].sum()

top_product_name = top_product.idxmax()

print("Most Sold Product:", top_product_name)


# --------------------------------------------------------------
# Q26: Check if there are loss making orders
# --------------------------------------------------------------

loss_orders = df[df["profit"] < 0]

print("Number of Loss Making Orders:", len(loss_orders))


# --------------------------------------------------------------
# Q27: Average sales per order
# --------------------------------------------------------------

avg_sales_order = df["sales"].mean()

print("Average Sales Per Order:", avg_sales_order)


# --------------------------------------------------------------
# FINAL BUSINESS INSIGHTS
# --------------------------------------------------------------

print("\n================ BUSINESS INSIGHTS ================")

# Insight 1
print("1. The category with the highest sales is:", highest_sales_category)

# Insight 2
print("2. The category generating the highest profit is:", highest_profit_category)

# Insight 3
print("3. The region contributing the most sales is:", top_region)

# Insight 4
print("4. The most profitable sub category is:", top_subcategory)

# Insight 5
print("5. The most sold product in the dataset is:", top_product_name)

# Insight 6
print("6. Total number of orders with loss:", len(loss_orders))

print("===================================================")
print("These insights help businesses understand which products,")
print("categories, and regions generate the most revenue and profit.")
```
